### 1.- Sistema de Ecuaciones

El sistema de ecuaciones a resolver es:



1. `x + 180p - 2160 = 0 => p = (2160 - x) / 180`
2. `300p - x - 2400 = 0 => p = (2400 + x) / 300`

```python
# Ecuaciones

def oferta(x):
    return (2400 + x) / 300

def demanda(x):
    return (2160 - x) / 180
```

### 2.- Encuentre el punto de equilibrio

```python
import sympy as sp

# Definimos las variables simbólicas
x, p = sp.symbols('x p')

# Definimos las ecuaciones del sistema
eq1 = sp.Eq(p, oferta(x))
eq2 = sp.Eq(p, demanda(x))

# Resolvemos el sistema
solucion = sp.solve((eq1, eq2), (x, p))

print(f"Solución encontrada: {solucion}")

# Filtramos por soluciones económicamente válidas (x > 0 y p > 0)
x_eq = solucion[x]
p_eq = solucion[p]
print(f"El punto de equilibrio es x = {x_eq} y p = {p_eq}")
```

```python
# Mostramos el resultado dinámicamente usando markdown
from IPython.display import Markdown, display

resultado_markdown = f"""
### Resultado

El sistema tiene dos soluciones posibles. Al filtrar por valores económicamente relevantes 
(donde tanto `x` como `p` son positivos), encontramos una única solución válida, 
por lo que el punto de equilibrio que satisface la ecuación es:

- **x = {x_eq} (cantidad)**
- **p = {p_eq} (precio)**

**Verificación:** 
- Oferta en x={x_eq}: p = {oferta(x_eq):.2f}
- Demanda en x={x_eq}: p = {demanda(x_eq):.2f}
"""

display(Markdown(resultado_markdown))
```

### 3.- Graficar las funciones

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(8, 6))
plt.figure(figsize=(8, 6))

#establecemos el rango de x
valores_x = np.linspace(0, 500, 100) # Dominio de la función, eje de las abcisas "x"

#graficamos las curvas
plt.plot(valores_x,
    oferta(valores_x),
    'blue',
    label='Oferta: $p = (2400 + x) / 300$' # Se colocan entre símbolo $ para que muestre la fórmula
    )
plt.plot(valores_x,
    demanda(valores_x),
    'red',
    label='Demanda: $(2160 - x) / 180$'
    )

#Agregamos los detalles del gráfico
plt.scatter(x_eq,
            p_eq,
            color='green',
            s=100, # tamaño en pixeles
            label=f'Equilibrio (x={x_eq:.2f}, p={p_eq:.2f})'
            )

plt.ylabel('Precio (p)')
plt.xlabel('Cantidad (x)')
plt.legend()
plt.grid(True)
plt.title('Equilibrio de Mercado ')
plt.show()
```

### 4- defina un hecho que modifique la demanda y otro hecho que modifique la oferta. Plantee que ambos se generan simultáneamente

Hecho que modifica la demanda:
Disminución en el ingreso de los consumidores, lo que desplaza la curva de demanda hacia la izquierda, se disminuye en 60 unidades

Hecho que modifica la oferta:
Aumento en los costos de producción, lo que desplaza la curva de oferta hacia la izquierda, se incrementa en 200 unidades


```python
# Funciones modificadas por hechos económicos

def oferta_modificada(x):
    return (2600 + x) / 300

def demanda_modificada(x):
    return (2100 - x) / 180


# Aplicamos ambos hechos simultáneamente
# Definimos las variables simbólicas
x, p = sp.symbols('x p')

# Definimos las ecuaciones del sistema
eq1_mod = sp.Eq(p, oferta_modificada(x))
eq2_mod = sp.Eq(p, demanda_modificada(x))

# Resolvemos el sistema
solucion = sp.solve((eq1_mod, eq2_mod), (x, p))

print(f"Solución encontrada: {solucion}")

# Filtramos por soluciones económicamente válidas (x > 0 y p > 0)
x_eq_mod = solucion[x]
p_eq_mod = solucion[p]
print(f"El punto de equilibrio es x = {x_eq_mod} y p = {p_eq_mod}")
```

## 5- explique los resultados

Al incrementar los costos, los oferentes están dispuestos a ofrecer menos productos con lo que la curva de oferta se desplaza hacia la izquierda. Esto provoca un aumento en el precio de equilibrio y una disminución en la cantidad de equilibrio. Por otro lado, la curva de demanda también se ve afectada, pero en este caso, se desplaza hacia la izquierda, lo que genera un nuevo punto de equilibrio. En resumen, los cambios en los costos de producción y en la demanda afectan directamente el equilibrio del mercado. En este caso se disminuyen las cantidades y se incrementa el precio
