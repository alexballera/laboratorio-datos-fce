# Parte 1

1. Crear una matriz de producción ficticia para una empresa que fabrica 3 productos en 2 plantas
2. Convertir la matriz en una tabla de datos
3. Calcular la producción total por planta (suma por filas)
4. Calcular la producción total por producto (suma por columnas)
5. Agregar columnas mediante operaciones que muestren los siguientes atributos: producción total por planta y participación porcentual de cada planta en la producción.
4. Calcular la proyección de una tercer planta que fabricará un 20% más de producto 1 , un 5% más de producto 2 y un 10% más de producto 3 que la primera planta  

```python
# Importaciones
import numpy as np
import pandas as pd
```

```python
# 1.-Crear una matriz de producción ficticia para una empresa que fabrica 3 productos en 2 plantas
# Crear arrays para cada planta
planta1_produccion = np.array([100, 150, 120])
planta2_produccion = np.array([200, 180, 250])

print("Producción de la Planta 1:")
print(planta1_produccion)

print("\nProducción de la Planta 2:")
print(planta2_produccion)

# Combinar los arrays en una matriz de producción
produccion_matriz = np.array([planta1_produccion, planta2_produccion])

print("\nMatriz de producción:")
print(produccion_matriz)
```

```python
# 2.-Convertir la matriz en una tabla de datos
df = pd.DataFrame(produccion_matriz, columns=['Producto 1', 'Producto 2', 'Producto 3'], index=['Planta 1', 'Planta 2'])
print(df)
```

```python
# 3.-Calcular la producción total por planta (suma por filas)
produccion_total_por_planta = df.sum(axis=1)
print("\nProducción total por planta:")
print(produccion_total_por_planta)
```

```python
# 4.-Calcular la producción total por producto (suma por columnas)
produccion_total_por_producto = df.sum(axis=0)
print("\nProducción total por producto:")
print(produccion_total_por_producto)
```

```python
# 5.-Agregar columnas mediante operaciones que muestren los siguientes atributos:
# producción total por planta y
# participación porcentual de cada planta en la producción

# Agregar la columna de producción total por planta
df['Producción Total por Planta'] = produccion_total_por_planta

# Calcular la participación porcentual de cada planta
produccion_total_general = produccion_total_por_planta.sum()
df['Participación Porcentual'] = (produccion_total_por_planta / produccion_total_general) * 100

print("\nDataFrame con producción total y participación porcentual:")
print(df)
```

```python
# 6.-Calcular la proyección de una tercer planta que fabricará un 20% más de producto 1,
# un 5% más de producto 2 y un 10% más de producto 3 que la primera planta

# Obtener la producción de la Planta 1
produccion_planta1 = df.loc['Planta 1', ['Producto 1', 'Producto 2', 'Producto 3']]

# Calcular la producción proyectada para la Planta 3
proyeccion_planta3 = produccion_planta1 * np.array([1.20, 1.05, 1.10])

print("\nProducción proyectada para la Planta 3:")
print(proyeccion_planta3)
```

# Parte 2 
1. Generar una matriz organizacional que contenga información sobre 5 productos incluyendo: cantidades vendidas, precios unitarios, costos unitarios. 
2. Transformar la matriz en una tabla de datos y agregar una columna que contenga los beneficios por producto 
3. Calcular beneficio total y el peso relativo del beneficio de cada producto sobre el total. 
4. Hacer los siguientes gráficos de visualización :
    1. Cantidades vendidas por producto, 
    2. Beneficio obtenido por producto 
    3. Relación entre precios de cada producto y cantidades vendidas 
    4. Participación porcentual de cada producto en el beneficio total.

```python
# 1. Generar una matriz organizacional que contenga información sobre 5 productos incluyendo: cantidades vendidas, precios unitarios, costos unitarios.
productos = np.array([
    [500, 20.0, 12.0],  # Producto 1: [cantidades vendidas, precio unitario, costo unitario]
    [300, 15.0, 9.0],   # Producto 2
    [400, 25.0, 15.0],  # Producto 3
    [600, 10.0, 6.0],   # Producto 4
    [350, 30.0, 18.0]   # Producto 5
])
```

```python
# 2. Transformar la matriz en una tabla de datos y agregar una columna que contenga los beneficios por producto
df_productos = pd.DataFrame(productos, 
                           columns=['Cantidades Vendidas', 'Precio Unitario', 'Costo Unitario'],
                           index=['Producto 1', 'Producto 2', 'Producto 3', 'Producto 4', 'Producto 5'])

# Calcular beneficios por producto (Beneficio = (Precio - Costo) * Cantidad)
beneficio_producto = (df_productos['Precio Unitario'] - df_productos['Costo Unitario']) * df_productos['Cantidades Vendidas']
df_productos['Beneficio por Producto'] = beneficio_producto

print("DataFrame de productos con beneficios:")
print(df_productos)
```

```python
# 3. Calcular beneficio total y el peso relativo del beneficio de cada producto sobre el total.

# Calcular beneficio total
beneficio_total = beneficio_producto.sum()
print(f"Beneficio total: ${beneficio_total:,.2f}")

# Calcular peso relativo (participación porcentual) de cada producto
peso_relativo = round((beneficio_producto / beneficio_total) * 100, 2)
df_productos['Peso Relativo (%)'] = peso_relativo

print("\nPeso relativo del beneficio de cada producto:")
print(peso_relativo)
```

```python
# 4. Hacer los siguientes gráficos de visualización
# Profe acá pude lograr lo de los gráficos con diferentes colores
# Hay que realizarlo con subplots y con axes se crea una matriz, y con las coordenadas
# se accede a cada gráfico

import matplotlib.pyplot as plt

# Configurar el estilo de las gráficas
plt.style.use('default')
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Análisis de Productos - Visualizaciones', fontsize=16)

# 1. Cantidades vendidas por producto
axes[0, 0].bar(df_productos.index, df_productos['Cantidades Vendidas'], color='skyblue')
axes[0, 0].set_title('Cantidades Vendidas por Producto')
axes[0, 0].set_ylabel('Cantidades')
axes[0, 0].tick_params(axis='x', rotation=45)

# 2. Beneficio obtenido por producto
axes[0, 1].bar(df_productos.index, df_productos['Beneficio por Producto'], color='lightgreen')
axes[0, 1].set_title('Beneficio Obtenido por Producto')
axes[0, 1].set_ylabel('Beneficio ($)')
axes[0, 1].tick_params(axis='x', rotation=45)

# 3. Relación entre precios de cada producto y cantidades vendidas
axes[1, 0].scatter(df_productos['Precio Unitario'], df_productos['Cantidades Vendidas'], 
                  s=100, color='coral', alpha=0.7)
axes[1, 0].set_title('Relación Precio vs Cantidades Vendidas')
axes[1, 0].set_xlabel('Precio Unitario ($)')
axes[1, 0].set_ylabel('Cantidades Vendidas')
# Agregar etiquetas a los puntos
for i, producto in enumerate(df_productos.index):
    axes[1, 0].annotate(producto, (df_productos['Precio Unitario'][i], df_productos['Cantidades Vendidas'][i]), 
                       xytext=(5, 5), textcoords='offset points', fontsize=9)

# 4. Participación porcentual de cada producto en el beneficio total
axes[1, 1].pie(df_productos['Peso Relativo (%)'], labels=df_productos.index, autopct='%1.1f%%', 
               colors=['gold', 'lightcoral', 'lightskyblue', 'lightgreen', 'lightpink'])
axes[1, 1].set_title('Participación Porcentual en Beneficio Total')

plt.tight_layout()
plt.show()
```
