# Derivada Parcial 😎

> Antes de avanzar con el siguiente caso primero aprendamos que es una derivada parcial, una herramienta muy util cuando debemos analizar la variacion de una variable respecto a otra.
>> Veamos un ejemplo:


🧠 ¿Qué es una derivada parcial?
Cuando tenés una función con varias variables, como:

$$f(x,y) = 5x^2 + 2y^2$$

La derivada parcial te dice cómo cambia la función cuando solo cambia una variable y las demás se mantienen constantes.

$$\frac{\partial f}{\partial x}=10x$$$$\frac{\partial f}{\partial y}=4y$$




```python
import sympy as sp

# Definir variables simbólicas
x, y = sp.symbols('x y')

# Definir la función
f = 5*x**2 + 2*y**2

# Derivadas parciales
df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)

# Mostrar resultados
print("Función: f(x, y) =", f)
print("Derivada parcial respecto a x (∂f/∂x):", df_dx)
print("Derivada parcial respecto a y (∂f/∂y):", df_dy)
```

Hace sentido con lo que veniamos hablando. Veamos otro ejemplo:
$$f(x,y) = x^2 * y^2$$

```python
# Definir variables simbólicas
x, y = sp.symbols('x y')

# Definir la función
f = x**2 * y**2

# Derivadas parciales
df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)

# Mostrar resultados
print("Función: f(x, y) =", f)
print("Derivada parcial respecto a x (∂f/∂x):", df_dx)
print("Derivada parcial respecto a y (∂f/∂y):", df_dy)
```

Sigamos:
$$f(q,l) = q^2 - l^2 + q + l$$

```python
# Definir variables simbólicas
q, l = sp.symbols('q l')

# Definir la función
f = q**2 - l**2 + q + l

# Derivadas parciales
df_dq = sp.diff(f, q)
df_dl = sp.diff(f, l)

# Mostrar resultados
print("Función: f(q, l) =", f)
print("Derivada parcial respecto a x (∂f/∂q):", df_dq)
print("Derivada parcial respecto a y (∂f/∂l):", df_dl)
```

# Caso 3: Duopolio 👉 Coca vs Manaos
image.png


En el país de las gaseosas, la hegemonía de Coca-Cola fue históricamente indiscutible. Su fórmula secreta, su distribución global y su marca la posicionaron como líder absoluto en el mercado de bebidas azucaradas. Pero un día, desde el conurbano profundo, emergió un competidor inesperado: Manaos. Económica, popular y nacional, se metió de lleno en la heladera de los consumidores argentinos.

Así se inicia una guerra económica entre dos titanes:

🔴 Coca-Cola, con mayor poder de marca, pero precios más altos.

🔵 Manaos, con una estrategia agresiva de precios bajos y volumen.

## 🧠 ¿Cómo compiten estas empresas?
Ambas saben que no pueden fijar el precio (eso lo determina el mercado), pero sí pueden decidir cuánto producir (Competencia a la ***cournot***) . Cada una elige su nivel de producción tratando de maximizar su beneficio, creyendo que la otra no va a cambiar su estrategia.



## ⚔️ ¿Qué es la competencia a la Cournot?
> La competencia a la Cournot es un modelo de competencia entre pocas empresas (duopolios u oligopolios) que deciden cuánto producir, no el precio.

📌 Supuestos clave:
*  Cada empresa elige su cantidad de producción.
*  Todas producen el mismo bien, que es homogéneo
*   El precio lo fija el mercado según la cantidad total producida.
*   Cada empresa cree que la cantidad del otro es fija y no reacciona.
*   Se decide simultáneamente de manera estrategica.








## 📈 ¿Qué vamos a analizar?
¿Qué cantidad produce cada empresa?


1. ¿Cuál es el precio que surge del mercado?

2.  ¿Cuánto gana cada una?

3.  ¿Cuál es el excedente del consumidor?

4. Y lo más importante: ¿cuál es el equilibrio que se genera cuando dos empresas compiten en cantidades?


---



La estructura de beneficios de las firmas responden a la siguiente forma : $$(50 - Q) Q - 10 Q$$

Donde $$Q = qA + qB$$

Siendo A Coca y B Manaos. Como bien podemos observar los costos son simetricos ,es decir, iguales para ambos

Arranquemos por coca $$(50 - qA - qB) qA - 10 qA$$

```python
def beneficio_A_duo(vars):
    qA, qB = vars
    return -((50 - qA - qB) * qA - 10*qA)

# Bounds para qA y qB
bounds = [(0, None), (0, None)]

# Inicial guess
x0 = [10, 10]

# Resolver
res_duo = minimize(beneficio_A_duo, x0=x0, bounds=bounds)
qA_opt = res_duo.x[0]
qB_opt = res_duo.x[1]
Q_opt2 = qA_opt + qB_opt
p_opt2 = 50 - Q_opt2

# Beneficio conjunto (solo para comparación)
benef_duopolio = (p_opt2 - 10) * Q_opt2

# Excedente del consumidor
EC_duo = 0.5 * (50 - p_opt2) * Q_opt2

print(f"Duopolio Cournot:")
print(f"  qA = {qA_opt:.2f}, qB = {qB_opt:.2f}")
print(f"  Q total = {Q_opt2:.2f}")
print(f"  Precio = {p_opt2:.2f}")
print(f"  Beneficio conjunto = {benef_duopolio:.2f}")
print(f"  Excedente consumidor = {EC_duo:.2f}")
```

En el modelo que resolvimos con ***scipy.optimize.minimize***, maximizamos únicamente el beneficio de la Empresa A, tratando la cantidad producida por la Empresa B como si fuera una variable más bajo su control.

👉 Esto viola un principio fundamental del modelo de Cournot: cada empresa decide su cantidad estratégicamente e independientemente, suponiendo fija la cantidad de la otra.
Al hacer que A elija tanto su cantidad como la de B, asumimos erróneamente que controla a su competidor.

Por eso, para modelar un verdadero equilibrio de Cournot, necesitamos una

herramienta simbólica como sympy, que nos permita:



*   Definir dos variables de decisión independientes (una por empresa),
*   Construir las dos funciones de beneficio por separado,
*   Y encontrar el equilibrio resolviendo el sistema de mejores respuestas simultáneas.










✅ Solo así obtenemos un modelo consistente con la competencia estratégica entre empresas, donde el equilibrio surge de la interacción mutua entre decisiones individuales, no de una única optimización conjunta.

```python
# Variables
qA, qB = sp.symbols('qA qB')

# Funciones de beneficio
benef_A = (50 - qA - qB)*qA - 10*qA
benef_B = (50 - qA - qB)*qB - 10*qB

# Derivadas
DEV_A = sp.diff(benef_A, qA)
DEV_B = sp.diff(benef_B, qB)

# Sistema de ecuaciones
sol = sp.solve([DEV_A, DEV_B], (qA, qB))

# Resultados
qA_star = sol[qA]
qB_star = sol[qB]
Q_star = qA_star + qB_star
P_star = 50 - Q_star
Beneficio_A= (50-Q_star)*qA_star - 10*qA_star
Beneficio_B= (50-Q_star)*qB_star - 10*qB_star
Beneficio_total= Beneficio_A + Beneficio_B
EC = (1/2)*(50 - P_star)*Q_star
B_social = EC + Beneficio_total

print(f"qA* = {qA_star}, qB* = {qB_star}")
print(f"Q* = {Q_star}, P* = {P_star}")
print(f"Beneficio A = {Beneficio_A}, Beneficio B = {Beneficio_B}")
print(f"Beneficio total = {Beneficio_total}")
print(f"Excedente del consumidor = {sp.simplify(EC)} ≈ {float(EC):.2f}")
print(f"Beneficio social = {sp.simplify(B_social)} ≈ {float(B_social):.2f}")
```

Obtenemos cuantas cantidades lleva cocacola y manaos al mercado, ambas llevar 13,33 millones de litros de gaseosa al mercado por mes. dando como resultado

```python
#generamos la D.inversa
q_vals = np.linspace(0, 50, 200)
p_vals = 50 - q_vals
plt.figure(figsize=(8,6))
plt.plot(q_vals, p_vals, label="Demanda", color='blue')

# pasamos sympy a float es decir numeros para que entienda matplot
Q_star_float = float(Q_star)
P_star_float = float(P_star)

plt.hlines(P_star_float, 0, Q_star_float, colors='green', linestyles='--', label=f'P* = {P_star_float:.2f}')
plt.vlines(Q_star_float, 0, P_star_float, colors='red', linestyles='--', label=f'Q* = {Q_star_float:.2f}')
plt.fill_between(q_vals[q_vals <= Q_star_float], p_vals[q_vals <= Q_star_float], P_star_float, color='cyan', alpha=0.3, label='Excedente consumidor')

plt.title("Equilibrio de Duopolio Cournot")
plt.xlabel("Cantidad total (Q = qA + qB)")
plt.ylabel("Precio (P)")
plt.legend()
plt.grid(True)
plt.show()
```

## ¿Y si manaos se queda con todo el mercado? 🥸
Ahora comparemos que sucederia si Manaos se queda con todo el mercado y pasa a ser monopolista.

La estructura de beneficios de las firmas responden a la siguiente forma : $$(50 - Q) Q - 10 Q$$

Donde $$Q = Qmanaos$$

```python
# Variables
Q = sp.symbols('Q')

# Funciones de beneficio
benef_Manaos = (50 - Q)*Q - 10*Q

# Derivadas
DEV_Q = sp.diff(benef_Manaos, Q)

# Sistema de ecuaciones
sol = sp.solve([DEV_Q], (Q))

# Resultados
Q_star = sol[Q]
P_star = 50 - Q_star
Beneficio_manaos= (50-Q_star)*Q_star - 10*Q_star
EC = (1/2)*(50 - P_star)*Q_star
B_social = EC + Beneficio_manaos

print(f"Q* = {Q_star}")
print(f"P* = {P_star}")
print(f"Beneficio Manaos = {Beneficio_manaos}")
print(f"Excedente del consumidor = {sp.simplify(EC)} ≈ {float(EC):.2f}")
print(f"Beneficio social = {sp.simplify(B_social)} ≈ {float(B_social):.2f}")
```

Como vemos el beneficio social disminuye y el excedente del consumidor tambien. La gente esta mejor con mas competencia por que pueden conseguir mas del producto a un precio mas barato.

Consecuencias del monopolio:

*   Ineficiencia asignativa: El monopolio produce menos cantidad y cobra un precio más alto que en competencia perfecta, lo que genera una pérdida de bienestar social (algunos consumidores dispuestos a pagar no pueden acceder al producto)

*   Ineficiencia productiva: Al no enfrentar competencia, el monopolista puede tener menos incentivos para minimizar costos o innovar.



*   Beneficios extraordinarios: El monopolista obtiene ganancias superiores a las normales debido a su capacidad de fijar precios.

¿Es siempre negativo el monopolio?
No necesariamente. En algunos casos, como los monopolios naturales (donde una sola empresa es más eficiente debido a altos costos fijos), la regulación puede asegurar que el monopolista no abuse de su poder. Además, los monopolios pueden incentivar la innovación si las empresas esperan obtener grandes beneficios (ejemplo: patentes en la industria farmacéutica).

# Caso de B: Optimización de Trabajo y Capital en la Industria Automotriz 🚗

Imaginemos a "AutoFuturo S.A.", una nueva empresa automotriz que busca establecer una línea de producción para un modelo de vehículo eléctrico innovador. Para ser competitivos, necesitan minimizar los costos de fabricación para alcanzar una meta de producción de 100 vehículos por día.

La producción de cada vehículo requiere una combinación de dos factores clave:

*   **Trabajo (L):** La mano de obra especializada (ingenieros, técnicos, operarios) que ensambla y supervisa el proceso. El costo de una unidad de trabajo (por ejemplo, horas de mano de obra calificada) es de \$10.
*   **Capital (K):** La maquinaria, robots, herramientas y tecnología utilizada en la línea de ensamblaje. El costo de una unidad de capital (por ejemplo, horas de uso de maquinaria avanzada) es de \$5.

La tecnología de producción de AutoFuturo S.A. se describe mediante la siguiente función de producción:

$$Q(L, K) = \sqrt{L \cdot K}$$

Donde:
*   $Q$ es la cantidad de vehículos producidos por día.
*   $L$ es la cantidad de unidades de trabajo utilizadas.
*   $K$ es la cantidad de unidades de capital utilizadas.

Esta función de producción, una forma de Cobb-Douglas, implica que tanto el trabajo como el capital son necesarios para la producción y que hay rendimientos a escala constantes (si duplicas tanto L como K, duplicas la producción).

AutoFuturo S.A. tiene como objetivo producir exactamente 100 vehículos por día. Por lo tanto, la restricción de producción es:

$$\sqrt{L \cdot K} = 100$$

O, elevando ambos lados al cuadrado:
$$L \cdot K = 10000$$

El costo total de producción ($C$) depende de las cantidades de trabajo ($L$) y capital ($K$) utilizadas y de sus respectivos precios ($w=10$ para el trabajo y $r=5$ para el capital):

$$C(L, K) = wL + rK = 10L + 5K$$


**El Desafío de AutoFuturo S.A.:** 🚗

El desafío para AutoFuturo S.A. es encontrar la combinación óptima de trabajo ($L^*$) y capital ($K^*$) que les permita producir los 100 vehículos diarios al menor costo total posible.

Este problema se resuelve utilizando técnicas de optimización restringida. Gráficamente, esto implica encontrar el punto donde una curva de isocosto (que representa todas las combinaciones de L y K que cuestan lo mismo) es tangente a la isocuanta (que representa todas las combinaciones de L y K que producen 100 unidades).

image.png

El objetivo es encontrar $L^*$ y $K^*$ tales que $10L^* + 5K^*$ sea mínimo, cumpliendo siempre la condición $\sqrt{L^* \cdot K^*} = 100$. La solución a este problema indicará a AutoFuturo S.A. cuántas unidades de trabajo y capital deben emplear para fabricar sus 100 vehículos eléctricos de la manera más eficiente en costos.

```python
# Función de costo
def costo(x):
    L, K = x
    return 10*L + 5*K  # Costo total

# Restricción: producir 100 unidades
def restriccion(x):
    L, K = x
    return (L * K)**0.5 - 100

# Llamamos al optimizador
res = minimize(costo, x0=[10, 10], constraints={'type': 'eq', 'fun': restriccion})

# Mostramos los resultados
L_opt, K_opt = res.x
print(f"L* = {L_opt:.2f}")
print(f"K* = {K_opt:.2f}")
print(f"Costo mínimo = {costo([L_opt, K_opt]):.2f}")
```

Perfect --> Sabemos las cantidades optimas de trabajo y capital que precisa la automotriz para producir 100 unidades de automoviles, arrojando un costo minimo de $1414,21

## Gráfico del caso 🙂


1.  **Calcular la Isocuanta:** Se utilizó la restricción de producción $L \cdot K = 10000$ (derivada de $\sqrt{L \cdot K} = 100$) para despejar $K$ en función de $L$. La ecuación de la isocuanta es $K = \frac{10000}{L}$.

2.  **Calcular la Isocosto Óptima:** Se utilizó el costo mínimo (`C_opt`) encontrado previamente por el optimizador. La ecuación general de una línea de isocosto es $C = wL + rK$. Despejando $K$: $K = \frac{C - wL}{r}$. Con $w=10$, $r=5$ y el costo mínimo `C_opt`, la ecuación de la isocosto óptima es $K = \frac{C_{opt} - 10L}{5}$. Se definió una función `isocosto(L, C)` para calcular estos valores de K para diferentes valores de L a un costo fijo C. Luego se graficó `L_vals` contra `isocosto(L_vals, C_opt)` para dibujar la línea de isocosto que pasa por el punto óptimo.

3.  **Marcar el Punto Óptimo:** Los valores óptimos de Trabajo (`L_opt`) y Capital (`K_opt`) obtenidos del resultado de la optimización (`res.x`) se graficaron como un punto rojo ('ro') para mostrar visualmente la combinación de L y K que minimiza el costo.


```python
# Rango para graficar
L_vals = np.linspace(1, L_opt + 10, 100)
K_vals_isoq = (100**2) / L_vals  # K = (Q^2) / L

# Función de isocosto
def isocosto(L, C): return (C - 10*L) / 5

# Gráfico
plt.figure(figsize=(8,6))
plt.plot(L_vals, K_vals_isoq, label='Isocuanta Q=100', color='blue')
plt.plot(L_vals, isocosto(L_vals, C_opt), label=f'Isocosto óptimo (C={C_opt:.0f})', color='green')

# Punto óptimo
plt.plot(L_opt, K_opt, 'ro', label='Óptimo')
plt.annotate(f'({L_opt:.1f}, {K_opt:.1f})',
             xy=(L_opt, K_opt), xytext=(L_opt+1, K_opt+20),
             arrowprops=dict(arrowstyle='->', color='red'), fontsize=10)

# Ajuste de ejes
plt.xlim(0, L_opt + 10)
plt.ylim(0, K_opt + 50)

plt.xlabel("Trabajo (L)")
plt.ylabel("Capital (K)")
plt.title("Minimización de Costos con Isocuanta e Isocosto")
plt.legend()
plt.grid(True)
plt.show()
```
