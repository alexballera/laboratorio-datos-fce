## 1. Importar librerías y cargar datos

Primero voy a importar las librerías que necesito para trabajar con datos.

```python
# Importar las librerías que voy a usar
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print("Librerías importadas correctamente")
```

```python
# Cargar los datos del archivo CSV
datos = pd.read_csv("actividad3.csv")

# Ver información básica
print("Datos cargados exitosamente")
print(f"Cantidad de propiedades: {datos.shape[0]}")
print(f"Cantidad de columnas: {datos.shape[1]}")

# Mostrar las primeras filas
datos.head()
```

## 2. Explorar los datos

Ahora voy a ver qué tienen los datos, calcular algunos promedios y contar cosas básicas.

```python
# Ver información general de los datos
print("INFORMACION BASICA DE LOS DATOS")
print("Total de propiedades:", len(datos))
print("Columnas disponibles:", list(datos.columns))

# Ver si hay valores faltantes
print("\nValores faltantes por columna:")
print(datos.isnull().sum())
```

### 2. Datos estadísticos básicos

Ahora veo los números básicos de los datos para entender mejor las propiedades.

```python
# Estadísticas básicas de los datos
print("ESTADISTICAS BASICAS")
print(datos.describe())

# Ver algunos datos específicos
print("\nAlgunos datos interesantes:")
print("Precio promedio:", datos['Price'].mean())
print("Precio mínimo:", datos['Price'].min())
print("Precio máximo:", datos['Price'].max())
```

```python
# Ver las zonas donde están las propiedades
print("ZONAS DONDE ESTAN LAS PROPIEDADES")
zonas = datos['Regionname'].value_counts()
print("Las 5 zonas con más propiedades:")
print(zonas.head())

# Ver el rango de fechas de los datos
print("\nFECHAS DE LOS DATOS")
print("Fecha más antigua:", datos['Date'].min())
print("Fecha más reciente:", datos['Date'].max())
```

### 3. Gráficos

Voy a hacer algunos gráficos simples para ver mejor los datos.

```python
# Gráfico 1: Histograma de precios
plt.figure(figsize=(10, 6))
plt.hist(datos['Price'], bins=30, color='blue', alpha=0.7)
plt.title('Distribución de Precios')
plt.xlabel('Precio')
plt.ylabel('Cantidad de propiedades')
plt.show()
```

```python
# Gráfico 2: Precios por zona
plt.figure(figsize=(12, 6))

# Precio promedio por zona
precio_por_zona = datos.groupby('Regionname')['Price'].mean().sort_values()
precio_por_zona.plot(kind='bar', color='green')
plt.title('Precio Promedio por Zona')
plt.xlabel('Zona')
plt.ylabel('Precio Promedio')
plt.xticks(rotation=45)
plt.show()
```

```python
# Gráfico 3: Precio vs Distancia al centro
plt.figure(figsize=(10, 6))
plt.scatter(datos['Distance'], datos['Price'], alpha=0.5)
plt.title('Precio vs Distancia al Centro')
plt.xlabel('Distancia (km)')
plt.ylabel('Precio')
plt.show()
```

### 4. Resumen de los datos

Voy a hacer un resumen simple de lo que encontré en los datos.

```python
# Resumen por tipo de propiedad
print("RESUMEN POR TIPO DE PROPIEDAD")
print("="*40)

# Para cada tipo de propiedad mostrar info básica
tipos = datos['Type'].unique()
for tipo in tipos:
    datos_tipo = datos[datos['Type'] == tipo]
    print(f"\nTipo {tipo}:")
    print(f"- Cantidad: {len(datos_tipo)}")
    print(f"- Precio promedio: ${datos_tipo['Price'].mean():.0f}")
    print(f"- Habitaciones promedio: {datos_tipo['Rooms'].mean():.1f}")
```

```python
# Resumen por zona
print("\nRESUMEN POR ZONA")
print("="*25)

# Mostrar las zonas con mayor y menor precio promedio
precio_por_zona = datos.groupby('Regionname')['Price'].mean().sort_values()

print("Zona más barata:", precio_por_zona.index[0], f"- ${precio_por_zona.iloc[0]:.0f}")
print("Zona más cara:", precio_por_zona.index[-1], f"- ${precio_por_zona.iloc[-1]:.0f}")
```

### 5. Conclusiones

Estas son mis conclusiones simples sobre los datos de propiedades.

```python
# Conclusiones simples
print("MIS CONCLUSIONES SOBRE LOS DATOS")
print("="*35)

print("1. Precio promedio de propiedades:", f"${datos['Price'].mean():.0f}")
print("2. La zona más cara es:", datos.groupby('Regionname')['Price'].mean().idxmax())
print("3. La zona más barata es:", datos.groupby('Regionname')['Price'].mean().idxmin())
print("4. Tipo de propiedad más común:", datos['Type'].mode()[0])
print("5. Número promedio de habitaciones:", f"{datos['Rooms'].mean():.1f}")
```
